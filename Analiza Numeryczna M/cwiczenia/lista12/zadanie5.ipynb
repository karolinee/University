```julia
using PyPlot
```

```julia
function kula(k)
    g = 9.81
    h = 0.01
    z(a,b) = k * sqrt((a^2) + (b^2))
    X = Float64[]
    Y = Float64[]
    fi = 60
    u0 = 100cosd(fi)
    v0 = 100sind(fi)
    x0 = 0
    y0 = 0
    u, v, x, y = u0, v0, x0, y0
    while y >= 0 
        push!(X, x)
        push!(Y, y)
        x = x + h*u        
        y = y + h*v       
        u_n = u - h * u * k * sqrt(u^2 + v^2)
        v = v - h * (g + v * k * sqrt(u^2 + v^2))
        u = u_n
    end
    return X, Y
end
```

```julia
xs0,ys0 = kula(0)
xs1,ys1 = kula(0.0001)
xs2, ys2 = kula(0.0003)
xs3, ys3 = kula(0.0005)
xs4, ys4 = kula(0.0007)

plot(xs0,ys0, xs1,ys1, xs2,ys2, xs3,ys3, xs4,ys4)

```
